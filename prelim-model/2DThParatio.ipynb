Solves the [Th]/[Pa] on a 2D grid using an upwind advection scheme, and contains plots of [Th]/[Pa] profile for a zero velocity field, a single cell velocity field, and a two cell velocity field. 

```python
from __future__ import division
import numpy
import pylab
import math
import matplotlib.pyplot as plt
from math import pi
import Th2D
%matplotlib inline
```

# [Th]/[Pa] in Zero Velocity Field

```python
# set parameters

T = 0.01
V = 1/100
xmin = 1
xmax = 1000000
zmin = 0
zmax = 5000
nz = 20
nx = 20

# velocity field
[u, flowfig] = Th2D.u_zero(xmin, xmax, zmin, zmax, nx, nz)

# Th chemistry
[Tk_ad, Tk_de, TQ] = Th2D.k_sorp('Th', xmin, xmax, zmin, zmax, nx, nz)

# Pa chemistry
[Pk_ad, Pk_de, PQ] = Th2D.k_sorp('Pa', xmin, xmax, zmin, zmax, nx, nz)

# [Th] loop
[flowfig, meshTh, DTh, PTh] = Th2D.adflow(T, V, u, nz, nx, Tk_ad, Tk_de, TQ, flowfig, 'Th')

# [Pa] loop
[flowfig, meshTh, DPa, PPa] = Th2D.adflow(T, V, u, nz, nx, Pk_ad, Pk_de, PQ, flowfig, 'Pa')

# plot ratio of [Th]/[Pa]
ThPa = Th2D.plotratio(DTh, DPa, PTh, PPa, xmin, xmax, zmin, zmax, nx, nz, T)
pylab.close(2)
pylab.close(3)
```
