Solves the [Th]/[Pa] on a 2D grid using an upwind advection scheme, and contains plots of [Th]/[Pa] profile for a zero velocity field, a single cell velocity field, and a two cell velocity field. 

```python
from __future__ import division
import numpy
import pylab
import math
import matplotlib.pyplot as plt
from math import pi
import ThPa2D
%matplotlib inline
```

# [Th]/[Pa] in Zero Velocity Field

```python
# set parameters
time = (0, 1, 2, 3, 5, 10)
t = time[0]
T = time[1]
V = 0.02/1000*365.*86400.  # first number in mm/s, corrections give needed value in m/yr ~ 630
xmin = 1
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nx = 20
ng = 1
g = ThPa2D.FDgrid(nx, nz, ng)
h = ThPa2D.FDgrid(nx, nz, ng)

# velocity field
[u, flowfig, init] = ThPa2D.u_zero(g, h, xmin, xmax, zmin, zmax, nx, nz, V, 'Pa')

# Th chemistry
[Tk_ad, Tk_de, TQ] = ThPa2D.k_sorp('Th', xmin, xmax, zmin, zmax, nx, nz)

# Pa chemistry
[Pk_ad, Pk_de, PQ] = ThPa2D.k_sorp('Pa', xmin, xmax, zmin, zmax, nx, nz)

# [Th] loop
[DTh, PTh] = ThPa2D.adflow(g, h, t, T, u, Tk_ad, Tk_de, TQ, ThPa2D.upwind)

# [Pa] loop
g = ThPa2D.FDgrid(nx, nz, ng)
h = ThPa2D.FDgrid(nx, nz, ng)
[DPa, PPa] = ThPa2D.adflow(g, h, t, T, u, Pk_ad, Pk_de, PQ, ThPa2D.upwind)

# plot ratio of [Th]/[Pa]
ThPa = ThPa2D.plotratio(DTh.a, DPa.a, PTh.a, PPa.a, xmin, xmax, zmin, zmax, nx, nz, T)
```
