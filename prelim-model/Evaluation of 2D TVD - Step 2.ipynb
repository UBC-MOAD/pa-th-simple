### Evaluation of 2D TVD : Step 2 (Fix ux so its not spikey)

```python
from __future__ import division
import numpy as np
import math
import copy
import matplotlib.pyplot as plt
from math import pi
import flow
import plot
import profile_plots as pp
%matplotlib inline
```

Set up grid, flow speed

```python
# Grid
xmin = 0
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nx = 30

# Flow Speed
V = 0.02/1000*365.*86400.
```

Now we import code from flow.onecell_cen and step through it

```python
# define velocity on square grid, then scale onto rectangular grid. 
# calculate radius
a = zmax
b = zmax
x = np.linspace(-a/2, a/2, nx)
z = np.linspace(-b/2, b/2, nz)
[xx, zz] = np.meshgrid(x, z)
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)
rr = np.sqrt(xx**2 + zz**2)

# plot radius
fig, ax = plt.subplots(1,1)
mesh = ax.pcolormesh(rr)
ax.invert_yaxis()
ax.set_title('rr')
cbar = fig.colorbar(mesh, ax=ax)
```

<i>Radius function looks like we expect

Continuing on we define idx, that part of the domain less than its size divided by two.

```python
idx = rr < a/2

# and plot it
cmap = plt.get_cmap('bwr')
fig, ax = plt.subplots(1,1)
mesh = ax.pcolormesh(idx, cmap=cmap)
ax.invert_yaxis()
ax.set_title('idx')
cbar = fig.colorbar(mesh, ax=ax)
```

<i>Looks as expected.  Note that one grid point all the way around is zero

Now define uz

```python
uz = np.zeros([nz, nx])
uz[idx] = np.sin(2*pi*rr[idx] / a) / rr[idx] * xx[idx]

# and plot it
fig, ax = plt.subplots(1,1)
mesh = ax.pcolormesh(uz, cmap=cmap)
ax.invert_yaxis()
ax.set_title('uz')
cbar = fig.colorbar(mesh, ax=ax)
```

<i>This looks nice.

Next section removes NaN, lets see where the NaN's are first

```python
fig, ax = plt.subplots(1,1)
mesh = ax.pcolormesh(np.isnan(uz), cmap=cmap)
ax.invert_yaxis()
ax.set_title('Nans in uz')
cbar = fig.colorbar(mesh, ax=ax)
```

<i>There are no Nans in uz.  This is because rr is never zero on this grid due to the grid spacing choice.  

Lets look at a different choice.

```python
# Grid
xmin = 0
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 21
nx = 31

# Flow Speed
V = 0.02/1000*365.*86400.

# define velocity on square grid, then scale onto rectangular grid. 
# calculate radius
a = zmax
b = zmax
x = np.linspace(-a/2, a/2, nx)
z = np.linspace(-b/2, b/2, nz)
[xx, zz] = np.meshgrid(x, z)
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)
rr = np.sqrt(xx**2 + zz**2)

# plot radius
fig, axs = plt.subplots(2,2,figsize=(10,10))
mesh=axs[0,0].pcolormesh(rr)
axs[0,0].invert_yaxis()
axs[0,0].set_title('rr')
cbar = fig.colorbar(mesh, ax=axs[0,0])

# find idx
idx = rr < a/2

# and plot it
cmap = plt.get_cmap('bwr')
mesh = axs[0,1].pcolormesh(idx, cmap=cmap)
axs[0,1].invert_yaxis()
axs[0,1].set_title('idx')
cbar = fig.colorbar(mesh, ax=axs[0,1])

# find uz
uz = np.zeros([nz, nx])
uz[idx] = np.sin(2*pi*rr[idx] / a) / rr[idx] * xx[idx]

# and plot it
mesh = axs[1,0].pcolormesh(uz, cmap=cmap)
axs[1,0].invert_yaxis()
axs[1,0].set_title('uz')
cbar = fig.colorbar(mesh, ax=axs[1,0])

# and plot Nans
mesh = axs[1,1].pcolormesh(np.isnan(uz), cmap=cmap)
axs[1,1].invert_yaxis()
ax.set_title('Nans in uz')
cbar = fig.colorbar(mesh, ax=axs[1,1])
```

<i> Okay now we have a Nan and that is causing the uz plot to fail

 However the function itself is well defined at zero.  Its goes to zero.  So lets just add a very small number to the denominator.

```python
vsmall = 1e-12
# find uz
uz = np.zeros([nz, nx])
uz[idx] = np.sin(2*pi*rr[idx] / a) / (rr[idx]+vsmall) * xx[idx]

# and plot it
fig, axs = plt.subplots(1,2,figsize=(10,5))
mesh = axs[0].pcolormesh(uz, cmap=cmap)
axs[0].invert_yaxis()
axs[0].set_title('uz')
cbar = fig.colorbar(mesh, ax=axs[0])

# and plot Nans
mesh = axs[1].pcolormesh(np.isnan(uz), cmap=cmap)
axs[1].invert_yaxis()
ax.set_title('Nans in uz')
cbar = fig.colorbar(mesh, ax=axs[1])
```

<i>All nice now.  And much easier than the remove Nans code.

Continue with the one cell code.  Scale uz and calculate ux.

```python
# scale uz
uz = uz / np.max(uz) * V * zmax/xmax

# calculate the grid by grid divergence of uz

div = np.zeros((nz,nx))
div[1:nz-1,:] = - dx/dz* ( uz[2:nz,:] - uz[0:nz-2,:])

# and plot it
fig, ax = plt.subplots(1,2)
mesh = ax[0].pcolormesh(div, cmap=cmap)
ax[0].invert_yaxis()
ax[0].set_title('div')
cbar = fig.colorbar(mesh, ax=ax[0])

# and do a line plot to look for spikiness
ax[1].plot(div[7,:],'*-')
ax[1].set_title('div across z=7')
```

<i> Okay this looks nice and smooth

Next calculate ux

```python
# set up ux
ux = np.zeros((nz,nx))

# integrate across to find ux
for j in range(2,nx):
    ux[:,j] = ux[:,j-2] + div[:,j-1]
    
# and plot it
fig, ax = plt.subplots(1,2,figsize=(10,5))
mesh = ax[0].pcolormesh(ux, cmap=cmap)
ax[0].invert_yaxis()
ax[0].set_title('ux')
cbar = fig.colorbar(mesh, ax=ax[0])

# and do a line plot to look for spikiness
ax[1].plot(ux[7,:],'*-')
ax[1].set_title('ux across z=7')
```

<i> So this is the origin of the spikiness.  It comes from the sharpness of the flow distribution at the start.  The divergence is very large at the start.

We can try a couple of things.  First, move the flow away from the boundaries.  I'm going to take this back to the grid step, so we go back to an even grid.

```python
# Grid
xmin = 0
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nx = 30

# Flow Speed
V = 0.02/1000*365.*86400.

# define velocity on square grid, then scale onto rectangular grid. 
# calculate radius
a = zmax
b = zmax
x = np.linspace(-a/2, a/2, nx)
z = np.linspace(-b/2, b/2, nz)
[xx, zz] = np.meshgrid(x, z)
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)
rr = np.sqrt(xx**2 + zz**2)

# plot radius
fig, axs = plt.subplots(2,2,figsize=(10,10))
mesh=axs[0,0].pcolormesh(rr)
axs[0,0].invert_yaxis()
axs[0,0].set_title('rr')
cbar = fig.colorbar(mesh, ax=axs[0,0])

# scale the flow in closer to centre of domain
scale = 0.9

# find idx
idx = rr < (a/2)*scale

# and plot it
cmap = plt.get_cmap('bwr')
mesh = axs[0,1].pcolormesh(idx, cmap=cmap)
axs[0,1].invert_yaxis()
axs[0,1].set_title('idx')
cbar = fig.colorbar(mesh, ax=axs[0,1])

# find uz
vsmall = 1e-12
uz = np.zeros([nz, nx])
uz[idx] = np.sin(2*pi*rr[idx] / (a*scale)) / (rr[idx]+vsmall) * xx[idx]

# and plot it
mesh = axs[1,0].pcolormesh(uz, cmap=cmap)
axs[1,0].invert_yaxis()
axs[1,0].set_title('uz')
cbar = fig.colorbar(mesh, ax=axs[1,0])

# and plot across 7
axs[1,1].plot(uz[7,:],'*-')
axs[1,1].set_title('uz across z=7')
```

<i>Ends are still too abrupt.  But a couple of zeros on outside is good.

Try two, lets also put a Gaussian on top to fade out the edges.

```python
# define velocity on square grid, then scale onto rectangular grid. 
# calculate radius
a = zmax
b = zmax
x = np.linspace(-a/2, a/2, nx)
z = np.linspace(-b/2, b/2, nz)
[xx, zz] = np.meshgrid(x, z)
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)
rr = np.sqrt(xx**2 + zz**2)

# plot radius
fig, axs = plt.subplots(2,2,figsize=(10,10))
mesh=axs[0,0].pcolormesh(rr)
axs[0,0].invert_yaxis()
axs[0,0].set_title('rr')
cbar = fig.colorbar(mesh, ax=axs[0,0])

# scale the flow in closer to centre of domain
scale = 0.9

# find idx
idx = rr < (a/2)*scale

# and plot it
cmap = plt.get_cmap('bwr')
mesh = axs[0,1].pcolormesh(idx, cmap=cmap)
axs[0,1].invert_yaxis()
axs[0,1].set_title('idx')
cbar = fig.colorbar(mesh, ax=axs[0,1])

# find uz
vsmall = 1e-12
uz = np.zeros([nz, nx])
uz[idx] = np.sin(2*pi*rr[idx] / (a*scale)) / (rr[idx]+vsmall) * xx[idx] 
uz_fade = np.zeros((nz, nx))
uz_fade[idx] = uz[idx] * np.exp(-4*rr[idx]**2/a**2) 

# and plot it
mesh = axs[1,0].pcolormesh(uz_fade, cmap=cmap)
axs[1,0].invert_yaxis()
axs[1,0].set_title('uz_fade')
cbar = fig.colorbar(mesh, ax=axs[1,0])

# and plot across 7
axs[1,1].plot(uz[7,:],'*-')
axs[1,1].plot(uz_fade[7,:],'ro-')
axs[1,1].set_title('uz and uz_fade across z=7')
```

<i> This looks less abrupt.  Lets test it.

Calculate the divergence, do plots

```python
uz = uz_fade
# scale uz
uz = uz / np.max(uz) * V * zmax/xmax

# calculate the grid by grid divergence of uz

div = np.zeros((nz,nx))
div[1:nz-1,:] = - dx/dz* ( uz[2:nz,:] - uz[0:nz-2,:])

# and plot it
fig, ax = plt.subplots(2,2, figsize=(7,7))
mesh = ax[0,0].pcolormesh(div, cmap=cmap)
ax[0,0].invert_yaxis()
ax[0,0].set_title('div')
cbar = fig.colorbar(mesh, ax=ax[0,0])

# and do a line plot to look for spikiness
ax[0,1].plot(div[7,:],'*-')
ax[0,1].set_title('div across z=7')

# set up ux
ux = np.zeros((nz,nx))

# integrate across to find ux
for j in range(2,nx):
    ux[:,j] = ux[:,j-2] + div[:,j-1]
    
# and plot it
mesh = ax[1,0].pcolormesh(ux, cmap=cmap)
ax[1,0].invert_yaxis()
ax[1,0].set_title('ux')
cbar = fig.colorbar(mesh, ax=ax[1,0])

# and do a line plot to look for spikiness
ax[1,1].plot(range(0,nx,2),ux[7,::2],'*-')
ax[1,1].plot(range(1,nx,2),ux[7,1::2],'o-r')
ax[1,1].set_title('ux across z=7')
```

<i> Much smoother.  It will be better if we have a grid spacing that is an integral of 4
