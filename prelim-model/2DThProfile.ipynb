Solves the [Th] on a 2D grid using an upwind advection scheme, and contains plots of [Th] profile for a zero velocity field, a single cell velocity field, and a two cell velocity field.

```python
from __future__ import division
import numpy as np
import pylab as plb
import copy
import matplotlib.pyplot as plt
from math import pi
import ThPa2D
import flow
import plot
%matplotlib inline
```

# [Th] for Zero Velocity Field

```python
# setup function adflow for Th 
reload(ThPa2D)
reload(flow)
reload(plot)
tmax = (0, 1, 2, 3, 5, 10, 20)
t = tmax[0]
T = tmax[1]
V = 0.02/1000*365.*86400.
xmin = 0
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nx = 30
ng = 1
g = ThPa2D.FDTgrid(nx, nz, ng)
h = ThPa2D.FPTgrid(nx, nz, ng)
# chemistry
k_ad, k_de, Q = ThPa2D.k_sorp('Th', xmin, xmax, zmin, zmax, nx, nz)
# velocity
u = flow.zero(nz, nx)
# initial profile
init, flowfig = plot.init(g.a, h.a, u, xmin, xmax, zmin, zmax, nx, nz, 'Th')
# Th loop
g, h = ThPa2D.adflow(g, h, t, T, u, k_ad, k_de, Q, ThPa2D.upstream_d, ThPa2D.upstream_p)
#plot results
meshTh = plot.prof(g, h, xmin, xmax, zmin, zmax, nx, nz, T, 'Th')
# store results
DTh0vel = {}
PTh0vel = {}
DTh0vel['1tmax'] = copy.copy(g.a)
PTh0vel['1tmax'] = copy.copy(h.a)
```
