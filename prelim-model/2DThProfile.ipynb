Solves the [Th] on a 2D grid using an upwind advection scheme, and contains plots of [Th] profile for a zero velocity field, a single cell velocity field, and a two cell velocity field.

```python
from __future__ import division
import numpy
import pylab
import math
import matplotlib.pyplot as plt
from math import pi
import ThPa2D
%matplotlib inline
```

# [Th] for Zero Velocity Field

```python
# setup function adflow for Th 

tmax = (0, 1, 2, 3, 5, 10)
t = tmax[0]
T = tmax[1]
V = 1/10
xmin = 1
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nx = 20
ng = 1
g = ThPa2D.FDgrid(nx, nz, ng)
h = ThPa2D.FDgrid(nx, nz, ng)

# chemistry
k_ad, k_de, Q = ThPa2D.k_sorp('Th', xmin, xmax, zmin, zmax, nx, nz)

# velocity
u, flowfig = ThPa2D.u_zero(g, h, xmin, xmax, zmin, zmax, nx, nz, 'Th')

# Th loop
g, h = ThPa2D.adflow(g, h, t, T, V, u, nz, nx, k_ad, k_de, Q)

#plot results
meshTh = ThPa2D.plotprof(g, h, xmin, xmax, zmin, zmax, nx, nz, T, 'Th')

# store results
DTh0vel = {}
PTh0vel = {}
DTh0vel['1tmax'] = g.a
PTh0vel['1tmax'] = h.a
```

```python
t = tmax[1]
T = tmax[2]
# Th loop
g, h = ThPa2D.adflow(g, h, t, T, V, u, nz, nx, k_ad, k_de, Q)

#plot results
meshTh = ThPa2D.plotprof(g, h, xmin, xmax, zmin, zmax, nx, nz, T, 'Th')

# store results
DTh0vel['2tmax'] = g.a
PTh0vel['2tmax'] = h.a
```

```python
t = tmax[2]
T = tmax[3]
# Th loop
g, h = ThPa2D.adflow(g, h, t, T, V, u, nz, nx, k_ad, k_de, Q)

#plot results
meshTh = ThPa2D.plotprof(g, h, xmin, xmax, zmin, zmax, nx, nz, T, 'Th')

# store results
DTh0vel['3tmax'] = g.a
PTh0vel['3tmax'] = h.a
```

```python
t = tmax[3]
T = tmax[4]
# Th loop
g, h = ThPa2D.adflow(g, h, t, T, V, u, nz, nx, k_ad, k_de, Q)

#plot results
meshTh = ThPa2D.plotprof(g, h, xmin, xmax, zmin, zmax, nx, nz, T, 'Th')

# store results
DTh0vel['5tmax'] = g.a
PTh0vel['5tmax'] = h.a
```

```python
t = tmax[4]
T = tmax[5]
# Th loop
g, h = ThPa2D.adflow(g, h, t, T, V, u, nz, nx, k_ad, k_de, Q)

#plot results
meshTh = ThPa2D.plotprof(g, h, xmin, xmax, zmin, zmax, nx, nz, T, 'Th')

# store results
DTh0vel['10tmax'] = g.a
PTh0vel['10tmax'] = h.a
```
