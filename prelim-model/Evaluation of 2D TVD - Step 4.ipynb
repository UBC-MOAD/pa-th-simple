Calculating flux in an upstream scheme breaks the symmetry of the profile.

Since the profile is initially constant, the flux is proportional to the velocity (which is dominantly x-velocity).

```python
# d(conc)/dt according to upstream scheme (on the grid points)
dtau_up_dt = np.zeros((nz, nx))
dtau_up_dt[1:nz, 1:nx] = (fluxx_up[1:nz, 0:nx-1] - fluxx_up[1:nz, 1:nx]) * conc.dx_i + (fluxz_up[0:nz-1, 1:nx] - fluxz_up[1:nz, 1:nx])  * conc.dz_i 
dtau_up_dt[0, 1:nx] = (fluxx_up[0, 0:nx-1] - fluxx_up[0, 1:nx]) * conc.dx_i - fluxz_up[0, 1:nx]  * conc.dz_i
dtau_up_dt[1:nz, 0] = -fluxx_up[1:nz, 0] * conc.dx_i + (fluxz_up[0:nz-1, 0] - fluxz_up[1:nz, 0])  * conc.dz_i
dtau_up_dt[0, 0] = - fluxx_up[0, 0] * conc.dx_i - fluxz_up[0, 0]  * conc.dz_i

# new concentration based on upstream scheme
tau_up = conc.a + dtau_up_dt * dt

# plot d(conc)/dt and conc(t+1)
fig = plt.subplots(1,3,figsize=(20,5))

plt.subplot(131)
test = plt.pcolormesh(dtau_up_dt)
plt.title('d(concentration)/dt based on upstream scheme')
plt.colorbar(test)

plt.subplot(132)
test = plt.pcolormesh(tau_up)
plt.title('new concentration based on upstream scheme')
plt.colorbar(test)

plt.subplot(133)
test = flow.divtest_cen(conc, u, nx, nz)
plt.title('divergence of velocity field')
```

Note the strange behaviour on RHS is due to wiggles in the velocity.

Why does the change in concentration have this pattern when the divergence of the velocity field is zero?

Try comparing with the change in concentration based on centred scheme:

```python
# centred flux
fluxx_cen = np.zeros((nz, nx));         fluxz_cen = np.zeros((nz, nx))
fluxx_cen[:, 0:nx-1] = 0.5 * ( conc.a[:, 0:nx-1]*ux[:, 0:nx-1] + conc.a[:, 1:nx]*ux[:, 1:nx] ) 
fluxz_cen[0:nz-1, :] = 0.5 * ( conc.a[0:nz-1, :]*(sinkrate + uz[0:nz-1, :]) + conc.a[1:nz, :]*(sinkrate + uz[1:nz, :] ))

# anti-diffusive flux
adfx = fluxx_cen - fluxx_up                     # conc*velocity
adfz = fluxz_cen - fluxz_up

# plot centred flux, anti-diffusive
fig, axs = plt.subplots(1,2,figsize=(15,5))

test = axs[0].pcolormesh(fluxx_cen)
axs[0].set_title('d(concentration)/dt based on centred scheme')
cbar = fig.colorbar(test, ax=axs[0])

test = axs[1].pcolormesh(fluxz_cen)
axs[1].set_title('new concentration based on centred scheme')
cbar = fig.colorbar(test, ax=axs[1])
```

```python
# define influx and outflux in x
xpos = np.zeros((nz, nx)); xneg = np.zeros((nz, nx))  
nfluxx = np.sign(adfx)*0.5*(np.sign(adfx) - 1)                          # dimensionless
pfluxx = np.sign(adfx)*0.5*(np.sign(adfx) + 1)
xpos[0:nz, 1:nx] = pfluxx[0:nz, 0:nx-1] * adfx[0:nz, 0:nx-1] - nfluxx[0:nz, 1:nx] * adfx[0:nz, 1:nx]    # conc*velocity
xpos[0:nz, 0] = - nfluxx[0:nz, 0] * adfx[0:nz, 0]
xneg[0:nz, 1:nx] = pfluxx[0:nz, 1:nx] * adfx[0:nz, 1:nx] - nfluxx[0:nz, 0:nx-1] * adfx[0:nz, 0:nx-1]
xneg[0:nz, 0] = pfluxx[0:nz, 0] * adfx[0:nz, 0] 

# influx and outflux in z 
zpos = np.zeros((nz, nx)); zneg = np.zeros((nz, nx))
nfluxz = np.sign(adfz)*0.5*(np.sign(adfz) - 1)
pfluxz = np.sign(adfz)*0.5*(np.sign(adfz) + 1)
zpos[1:nz, 0:nx] = pfluxz[0:nz-1, 0:nx] * adfz[0:nz-1, 0:nx] - nfluxz[1:nz, 0:nx] * adfz[1:nz, 0:nx]    # conc*velocity
zpos[0, 0:nx] = - nfluxz[0, 0:nx] * adfz[0, 0:nx]
zneg[1:nz, 0:nx] = pfluxz[1:nz, 0:nx] * adfz[1:nz, 0:nx] - nfluxz[0:nz-1, 0:nx] * adfz[0:nz-1, 0:nx]
zneg[0, 0:nx] = pfluxz[0, 0:nx] * adfz[0, 0:nx]

# total influx/outflux
fpos = xpos/conc.dz + zpos/conc.dx              # units: concentration/time
fneg = xneg/conc.dz + zneg/conc.dx

# plot 
fig, axs = plt.subplots(1,3,figsize=(20,5))

test = axs[0].pcolormesh(fpos)
axs[0].set_title('influx')
cbar = fig.colorbar(test, ax=axs[0])

test = axs[1].pcolormesh(fneg)
axs[1].set_title('out-flux')
cbar = fig.colorbar(test, ax=axs[1]) 

test = axs[2].pcolormesh(fpos - fneg)
axs[2].set_title('fpos - fneg')
cbar = fig.colorbar(test, ax=axs[2]) 

```

anti-diffusive flux is not necessarily conservative

```python
# non dimensional Zalesak parameter (produces nans when commented part is uncommented)
vsmall = 1e-12
# = (max_conc - upstream_conc) / influx
betaup = (conc_up - tau_up) / (fpos*dt + vsmall)
# = (upstream_conc - min_conc) / outflux
betado = (tau_up - conc_do) / (fneg*dt + vsmall)

# plot 
fig, axs = plt.subplots(1,2,figsize=(20,5))
test = axs[0].pcolormesh(betaup)
axs[0].set_title('betaup')
cbar = fig.colorbar(test, ax=axs[0])

test = axs[1].pcolormesh(betado)
axs[1].set_title('betado')
cbar = fig.colorbar(test, ax=axs[1])
```
