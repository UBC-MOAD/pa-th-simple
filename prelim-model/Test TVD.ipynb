```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Set-up a simple, constant flow u (from left to right), and a distribution of tracer (linearly increasing)

```python
# set up
nt = 10
dt = 0.5                                # s (maximum for stability for the central scheme)
nx = 11                                 
dx = 1.0                                # m
Pa = np.zeros((nt,nx))                  # conc units
Pa[0,:] = np.arange(nx)*1/(nx-1)        # conc units
u = np.ones((nt,nx))                    # m/s

# plot
maxc = 1.0                              # conc units
minc = 0.0
fig, ax = plt.subplots(1,1)
ax.plot(Pa[0])
ax.set_xlabel('x [m]')
ax.set_ylabel('Pa conc [normalized]')
ax.set_title('Initial Concentration')
ax.set_ylim(minc,maxc);
```

Next, calculate the flux based on the upstream method.   
I have u points at 0, 1, 2, etc.  Therefore calculate flux at 0.5, 1.5 etc (which I will call 0, 1 etc)   
flux_up === zwx at line 138 in traadv_trv.F90

and calculate the new concentration based on the upstream method
tau_up === zwi at line 162 (note pta includes source/sink terms, that ptb does not)

```python
# set up: upstream flux
flux_up = np.zeros(nx-1)
flux_up = u[0,:-1]*Pa[0,:-1]      # C m/s  (where C is conc units)

# plot
fig, axs = plt.subplots(1,2,figsize=(7,5))
axs[0].plot(np.arange(nx-1)+0.5,flux_up)
axs[0].set_xlabel('x [m]')
axs[0].set_ylabel('Flux [C m/s]')
axs[0].set_title('Flux using Upstream Method')
axs[0].set_ylim(minc,maxc);

# new concentration based on upstream method
tau_up = np.zeros(nx)
tau_up[0] = 0.
tau_up[1:-1] = Pa[0,1:-1]+(flux_up[0:-1]-flux_up[1:])/dx*dt           # C
tau_up[-1] = 2*tau_up[-2]-tau_up[-3]

# plot
axs[1].plot(tau_up,label='up')
axs[1].set_xlabel('x [m]')
axs[1].set_ylabel('Pa conc [C]')
axs[1].set_title('Pa at next time step using upstream')
axs[1].set_ylim(minc,maxc);
```

Calculate the centred flux estimate.   
Note that I don't see this saved anywhere.  But **adf is the anti-diffusive flux,  the difference between the centred scheme and the upstream scheme.**   
adf == zwx at line 186

```python
# central flux
flux_cen = np.zeros(nx-1)
flux_cen = 0.5*(Pa[0,:-1]*u[0,:-1]+Pa[0,1:]*u[0,1:])                # C m/s

# plot
fig, axs = plt.subplots(1,3,figsize=(12,5))
axs[0].plot(np.arange(nx-1)+0.5,flux_up,label='up')
axs[0].plot(np.arange(nx-1)+0.5,flux_cen,label='cen')
axs[0].set_xlabel('x [m]')
axs[0].set_ylabel('Flux [C m/s]')
axs[0].set_title('Flux using Central/Up Method')
axs[0].set_ylim(minc,maxc)
axs[0].legend(loc='lower right')

# anti-diffusive flux
adf = flux_cen - flux_up                                            # C m/s

# plot
axs[1].plot(np.arange(nx-1)+0.5,adf)
axs[1].set_xlabel('x [grid points]')
axs[1].set_ylabel('Flux')
axs[1].set_title('Anti-diffusive Flux')
axs[1].set_ylim((0.,0.1));

# just 'cause, new concentration based on central
tau_cen = np.zeros(nx)
tau_cen[0] = 0.
tau_cen[1:-1] = Pa[0,1:-1]+(flux_cen[0:-1]-flux_cen[1:])*dx/dt         # C
tau_cen[-1] = 2*tau_cen[-2]-tau_cen[-3]

# plot
axs[2].plot(tau_up,label='up')
axs[2].plot(tau_cen,label='cen')
axs[2].set_xlabel('x [m]')
axs[2].set_ylabel('Pa conc [C]')
axs[2].set_title('Pa at next time step using upstream/central')
axs[2].set_ylim(-0.2,maxc)
axs[2].legend(loc='lower right');
```

Find zup (maximum value in region) and zmin(minimum value in region)

```python
%%timeit
# calculate zup and zdo: max and min concentrations in region
zup = np.zeros(nx); zdo = np.zeros(nx); Pa3=np.zeros((3,nx)); t3=np.zeros((3,nx))
Pa3[0,1:nx-1] = Pa[0,0:nx-2]
Pa3[1,1:nx-1] = Pa[0,1:nx-1]
Pa3[2,1:nx-1] = Pa[0,2:nx]
t3[0,1:nx-1] = tau_up[0:nx-2]
t3[1,1:nx-1] = tau_up[1:nx-1]
t3[2,1:nx-1] = tau_up[2:nx]
zup = np.maximum(np.amax(Pa3,axis=0),np.amax(t3,axis=0))   # C
zdo = np.minimum(np.amin(Pa3,axis=0),np.amin(t3,axis=0))    # C

zup[0] = zup[1]; zdo[0] = zdo[1]
zup[-1] = zup[-2]; zdo[-1] = zdo[-2]

```

```python
%%timeit
# calculate zup and zdo: max and min concentrations in region
zup = np.zeros(nx); zdo = np.zeros(nx)
for i in range(1,nx-1):
    zup[i] = max(np.max(Pa[0,i-1:i+2]),np.max(tau_up[i-1:i+2]))    # C
    zdo[i] = min(np.min(Pa[0,i-1:i+2]),np.min(tau_up[i-1:i+2]))    # C
zup[0] = zup[1]; zdo[0] = zdo[1]
zup[-1] = zup[-2]; zdo[-1] = zdo[-2]
```

Find the positive and negative parts of the flux in and out of a point. (So on a tracer cell)

Following cell: the truth value is unambiguous because only on j location.

```python
# calculate positive and negative ad-fluxes
zpos = np.zeros(nx); zneg = np.zeros(nx)
for i in range(1,nx-1):
    
    zpos[i] = max(adf[i-1], 0.) - min(adf[i], 0)                      # C m/s
    zneg[i] = max(adf[i], 0) - min(adf[i-1], 0)                      # C m/s

    # guess some boundary conditions
zpos[0] = zpos[1]; zneg[0] = zneg[1];
zpos[-1] = zpos[-2]; zneg[-1] = zneg[-2]

# plot
fig, ax = plt.subplots(1,1)
ax.plot(zpos,'r',zneg,'b')
ax.set_ylabel('Flux [C m/s]')
ax.set_xlabel('x [m]')
ax.set_title("Positive and Negative Parts of the Anti-diffusive Flux")
ax.set_ylim(0,0.1);
```

Next are the up and down beta terms, where beta is the name used in NEMO.   
This scheme is the alternative flux limiter proposed by  
Zalesak, S.T.: Fully multidimensional Flux-Corrected Transport algorithms for fluids. J. Comput. Phys. 31, 335–362 (1979)   
which is the reference given by NEMO.  Its in the review paper Zalesak, 2012, see page 39 section 4.2   
There, zpos = P+ and zneg = P-


Q+, the numerator of zbetup, is the difference between max reasonable value (zup) and the upstream scheme (tau_up)

zbetup = Q+/P+


```python
zbetup = (zup - tau_up) / (zpos+0.0001) * dx/dt                       # C / (C m/s) * m/s = non dimensional
zbetdo = (tau_up - zdo) / (zneg+0.0001) * dx/dt

# plot
fig, axs = plt.subplots(1,2,figsize=(7,5))
axs[0].plot(zup,label='zup')
axs[0].plot(zdo,label='zdo')
axs[0].plot(tau_up,label='tau_up')
axs[0].legend(loc='lower right')
axs[1].plot(zbetup,'bo-',label='zbetup')
axs[1].plot(zbetdo,'ro-',label='zbetdo')
axs[1].legend(loc='lower right');
```

zcu = 1 if adf[i] > 0 and equals 0 if adf[i] <= 0


So if adf[i] > 0 we use zau 


and if adf[i] <= 0 we use zbu


Now zbu is the minimum of 1 and zbetup[i] and zbetdo[i+1].  This is equivalent to the min(1,R+,R-[i+1]) as spec'd in the Zalesak paper.

```python
# calculate zau, zbu and zcu
zau = np.zeros(nx-1); zbu = np.zeros(nx-1); zcu = np.zeros(nx-1)
for i in range(nx-1):
    zau[i] = min(1, zbetdo[i], zbetup[i+1])                            # non dim and on the flux points
    zbu[i] = min(1, zbetup[i], zbetdo[i+1])                            # non dim and on the flux points
    zcu[i] = (0.5 + 0.5*np.sign(adf[i]))                               # non dim and on the flux points

# plot
fig, axs = plt.subplots(1,2,figsize=(8,5))
axs[0].plot(np.arange(nx-1)+0.5,zau,'b*-',label='zau')
axs[0].plot(np.arange(nx-1)+0.5,zbu,'r*-',label='zbu')
axs[0].plot(np.arange(nx-1)+0.5,zcu,'g*-',label='zcu')
axs[0].set_xlabel('x [grid points]')
axs[0].set_ylim(minc,maxc);
axs[0].legend(loc='lower right')

# calculate final flux
paa = adf * (zcu * zau + (1-zcu)*zbu)                                # C m/s on flux points

axs[1].plot(np.arange(nx-1)+0.5,paa,label='paa')
axs[1].plot(np.arange(nx-1)+0.5,adf,label='adf')
axs[0].legend(loc='lower right');
```

Now calculate the final solution.  Note in NEMO pta has already been calculated based on the upstream method.

```python
# final solution
Pa[1,1:-1] = tau_up[1:-1] + (paa[0:-1] - paa[1:]) * dt/dx
Pa[1,-1] = 2*Pa[1,-2]-Pa[1,-3]
Pa[1,0] = 0

# plot
fig, ax = plt.subplots(1,1)
ax.plot(Pa[0],'b',Pa[1],'rx-')
ax.set_ylabel('Pa Concentration [C]')
ax.set_xlabel('x [m]')
ax.set_title("Original and time-stepped solution")
ax.set_ylim(minc,maxc);
```

Now strip and loop

```python
# anti-diffusive flux calculation
def limited_adf(adf,flux_up,tau_up,Pa):
    # calculate zup and zdo
    zup = np.zeros(nx); zdo = np.zeros(nx)
    for i in range(1,nx-1):
        zup[i] = max(np.max(Pa[i-1:i+2]),np.max(tau_up[i-1:i+1]))    # C
        zdo[i] = min(np.min(Pa[i-1:i+2]),np.min(tau_up[i-1:i+1]))    # C
    zup[0] = zup[1]; zdo[0] = zdo[1]
    zup[-1] = zup[-2]; zdo[-1] = zdo[-2]
    # calculate positive and negative fluxes
    zpos = np.zeros(nx); zneg = np.zeros(nx)
    for i in range(1,nx-1):
        zpos[i] = max(adf[i-1],0.)-min(adf[i],0)                      # C m/s
        zneg[i] = max(adf[i],0)-min(adf[i-1],0)                      # C m/s
    # guess some boundary conditions
    zpos[0] = zpos[1]; zneg[0] = zneg[1];
    zpos[-1] = zpos[-2]; zneg[-1] = zneg[-2]
    # beta terms
    zbetup = (zup - tau_up) / (zpos+0.0001) * dx/dt        # C / (C m/s) * m/s = none dimensional
    zbetdo = (tau_up - zdo) / (zneg+0.0001) * dx/dt
    # calculate zau, zbu and zcu
    zau = np.zeros(nx-1); zbu = np.zeros(nx-1); zcu = np.zeros(nx-1)
    for i in range(nx-1):
        zau[i] = min(1, zbetdo[i], zbetup[i+1])                 # non dim and on the flux points
        zbu[i] = min(1, zbetup[i], zbetdo[i+1])                 # non dim and on the flux points
        zcu[i] = (0.5 + 0.5*np.sign(adf[i]))                    # non dim and on the flux points
    # calculate final flux
    paa = adf * (zcu * zau + (1-zcu)*zbu)                                # C m/s on flux points
    return paa
```

```python
# loop over some time steps
for it in np.arange(2,6):
    # upstream flux
    flux_up = u[0,:-1]*Pa[it-1,:-1]
    # upstream solution
    tau_up[1:-1] = Pa[it-1,1:-1]+(flux_up[0:-1]-flux_up[1:])/dx*dt           # C
    # boundary conditions
    tau_up[-1] = 2*tau_up[-2]-tau_up[-3]; tau_up[0] = 0.
    # anti-diffusive fluxes
    flux_cen = 0.5*(Pa[it-1,:-1]*u[0,:-1]+Pa[it-1,1:]*u[0,1:])                # C m/s
    adf = flux_cen - flux_up                                            # C m/s
    # calculate limited anti-diffusive fluxes
    paa = limited_adf(adf,flux_up,tau_up,Pa[it-1])
    # final solution
    Pa[it,1:-1] = tau_up[1:-1] + (paa[0:-1] - paa[1:]) * dt/dx
    # boundary conditions
    Pa[it,-1] = 2*Pa[it,-2]-Pa[it,-3]; Pa[it,0] = 0

# plot
fig, ax = plt.subplots(1,1)
for it in np.arange(6):
    ax.plot(Pa[it],'o-')
ax.set_ylabel('Pa Concentration [C]')
ax.set_xlabel('x [m]')
ax.set_title("Original and time-stepped solution")
ax.set_ylim(minc,maxc);
```
