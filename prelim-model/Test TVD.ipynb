Find the positive and negative parts of the flux in and out of a point.  (So on a tracer cell)

Next are the up and down beta terms, where beta is the name used in NEMO.   
This scheme is the alternative flux limiter proposed by  
Zalesak, S.T.: Fully multidimensional Flux-Corrected Transport algorithms for fluids. J. Comput. Phys. 31, 335–362 (1979)   
which is the reference given by NEMO.  Its in the review paper Zalesak, 2012, see page 39 section 4.2   
There, zpos = P+ and zneg = P-


Q+, the numerator of zbetup, is the difference between max reasonable value (zup) and the upstream scheme (tau_up)

zbetup = Q+/P+


```python
zbetup = (zup - tau_up) / (zpos+0.0001) * dx/dt                       # C / (C m/s) * m/s = non dimensional
zbetdo = (tau_up - zdo) / (zneg+0.0001) * dx/dt

# plot
fig, axs = plt.subplots(1,2,figsize=(7,5))
axs[0].plot(zup,label='zup')
axs[0].plot(zdo,label='zdo')
axs[0].plot(tau_up,label='tau_up')
axs[0].legend(loc='lower right')
axs[1].plot(zbetup,'bo-',label='zbetup')
axs[1].plot(zbetdo,'ro-',label='zbetdo')
axs[1].legend(loc='lower right');
```

zcu = 1 if adf[i] > 0 and equals 0 if adf[i] <= 0


So if adf[i] > 0 we use zau 


and if adf[i] <= 0 we use zbu


Now zbu is the minimum of 1 and zbetup[i] and zbetdo[i+1].  This is equivalent to the min(1,R+,R-[i+1]) as spec'd in the Zalesak paper.

```python
# calculate zau, zbu and zcu
zau = np.zeros(nx-1); zbu = np.zeros(nx-1); zcu = np.zeros(nx-1)
for i in range(nx-1):
    zau[i] = min(1, zbetdo[i], zbetup[i+1])                            # non dim and on the flux points
    zbu[i] = min(1, zbetup[i], zbetdo[i+1])                            # non dim and on the flux points
    zcu[i] = (0.5 + 0.5*np.sign(adf[i]))                               # non dim and on the flux points

# plot
fig, axs = plt.subplots(1,2,figsize=(8,5))
axs[0].plot(np.arange(nx-1)+0.5,zau,'b*-',label='zau')
axs[0].plot(np.arange(nx-1)+0.5,zbu,'r*-',label='zbu')
axs[0].plot(np.arange(nx-1)+0.5,zcu,'g*-',label='zcu')
axs[0].set_xlabel('x [grid points]')
axs[0].set_ylim(minc,maxc);
axs[0].legend(loc='lower right')

# calculate final flux
paa = adf * (zcu * zau + (1-zcu)*zbu)                                # C m/s on flux points

axs[1].plot(np.arange(nx-1)+0.5,paa,label='paa')
axs[1].plot(np.arange(nx-1)+0.5,adf,label='adf')
axs[0].legend(loc='lower right');
```

Now calculate the final solution.  Note in NEMO pta has already been calculated based on the upstream method.

```python
# final solution
Pa[1,1:-1] = tau_up[1:-1] + (paa[0:-1] - paa[1:]) * dt/dx
Pa[1,-1] = 2*Pa[1,-2]-Pa[1,-3]
Pa[1,0] = 0

# plot
fig, ax = plt.subplots(1,1)
ax.plot(Pa[0],'b',Pa[1],'rx-')
ax.set_ylabel('Pa Concentration [C]')
ax.set_xlabel('x [m]')
ax.set_title("Original and time-stepped solution")
ax.set_ylim(minc,maxc);
```

```python
# anti-diffusive flux calculation
def limited_adf(adf,flux_up,tau_up,Pa):
    # calculate zup and zdo
    zup = np.zeros(nx); zdo = np.zeros(nx)
    for i in range(1,nx-1):
        zup[i] = max(np.max(Pa[i-1:i+2]),np.max(tau_up[i-1:i+1]))    # C
        zdo[i] = min(np.min(Pa[i-1:i+2]),np.min(tau_up[i-1:i+1]))    # C
    zup[0] = zup[1]; zdo[0] = zdo[1]
    zup[-1] = zup[-2]; zdo[-1] = zdo[-2]
    # calculate positive and negative fluxes
    zpos = np.zeros(nx); zneg = np.zeros(nx)
    for i in range(1,nx-1):
        zpos[i] = max(adf[i-1],0.)-min(adf[i],0)                      # C m/s
        zneg[i] = max(adf[i],0)-min(adf[i-1],0)                      # C m/s
    # guess some boundary conditions
    zpos[0] = zpos[1]; zneg[0] = zneg[1];
    zpos[-1] = zpos[-2]; zneg[-1] = zneg[-2]
    # beta terms
    zbetup = (zup - tau_up) / (zpos+0.0001) * dx/dt        # C / (C m/s) * m/s = none dimensional
    zbetdo = (tau_up - zdo) / (zneg+0.0001) * dx/dt
    # calculate zau, zbu and zcu
    zau = np.zeros(nx-1); zbu = np.zeros(nx-1); zcu = np.zeros(nx-1)
    for i in range(nx-1):
        zau[i] = min(1, zbetdo[i], zbetup[i+1])                 # non dim and on the flux points
        zbu[i] = min(1, zbetup[i], zbetdo[i+1])                 # non dim and on the flux points
        zcu[i] = (0.5 + 0.5*np.sign(adf[i]))                    # non dim and on the flux points
    # calculate final flux
    paa = adf * (zcu * zau + (1-zcu)*zbu)                                # C m/s on flux points
    return paa
```

```python
# loop over some time steps
for it in np.arange(2,6):
    # upstream flux
    flux_up = u[0,:-1]*Pa[it-1,:-1]
    # upstream solution
    tau_up[1:-1] = Pa[it-1,1:-1]+(flux_up[0:-1]-flux_up[1:])/dx*dt           # C
    # boundary conditions
    tau_up[-1] = 2*tau_up[-2]-tau_up[-3]; tau_up[0] = 0.
    # anti-diffusive fluxes
    flux_cen = 0.5*(Pa[it-1,:-1]*u[0,:-1]+Pa[it-1,1:]*u[0,1:])                # C m/s
    adf = flux_cen - flux_up                                            # C m/s
    # calculate limited anti-diffusive fluxes
    paa = limited_adf(adf,flux_up,tau_up,Pa[it-1])
    # final solution
    Pa[it,1:-1] = tau_up[1:-1] + (paa[0:-1] - paa[1:]) * dt/dx
    # boundary conditions
    Pa[it,-1] = 2*Pa[it,-2]-Pa[it,-3]; Pa[it,0] = 0

# plot
fig, ax = plt.subplots(1,1)
for it in np.arange(6):
    ax.plot(Pa[it],'o-')
ax.set_ylabel('Pa Concentration [C]')
ax.set_xlabel('x [m]')
ax.set_title("Original and time-stepped solution")
ax.set_ylim(minc,maxc);
```

```python
#def TVD(conc, u, p_upz, n_upz, p_upx, n_upx, sinkrate):
"""total variance diminishing advection scheme
"""
conc = g
# define grid
nz = conc.nz
nx = conc.nx
dt = 0.001
# extract velocity
uz = u[0,:,:]
ux = u[1,:,:]

# define upstream flux (in multi-directional flux field)
fluxx_up = np.zeros((nz, nx))
fluxx_up[1:nz-2, 1:nx-2] = ux[1:nz-2, 0:nx-3]*conc.a[1:nz-2, 0:nx-3] * n_upx[1:nz-2, 1:nx-2] + ux[1:nz-2, 2:nx-1]*conc.a[1:nz-2, 2:nx-1] * p_upx[1:nz-2, 1:nx-2]
fluxz_up = np.zeros((nz, nx))
fluxz_up[1:nz-2, 1:nx-2] = uz[0:nz-3, 1:nx-2]*conc.a[0:nz-3, 1:nx-2] * n_upz[1:nz-2, 1:nx-2] + uz[2:nz-1, 1:nx-2]*conc.a[2:nz-1, 1:nx-2] * p_upz[1:nz-2, 1:nx-2]

# d(conc)/dt based on upstream-flux method
tau_up = np.zeros((nz, nx))
tau_up[1:nz-2, 1:nx-2] = ((fluxx_up[1:nz-2, 0:nx-3] - fluxx_up[1:nz-2, 1:nx-2]) * n_upx[1:nz-2, 1:nx-2] + (fluxx_up[1:nz-2, 1:nx-2] - fluxx_up[1:nz-2, 2:nx - 1]) * p_upx[1:nz-2, 1:nx-2]) * conc.dx_i + ((fluxz_up[0:nz-3, 1:nx-2] - fluxz_up[1:nz-2, 1:nx-2]) * n_upz[1:nz-2, 1:nx-2] + (fluxz_up[1:nz-2, 1:nx-2] - fluxz_up[2:nz - 1, 1:nx-2]) * p_upz[1:nz-2, 1:nx-2]) * conc.dz_i 

# define centred flux
fluxx_cen = np.zeros((nz, nx))
fluxx_cen[1:nz-2, 1:nx-2] = 0.5 * ( conc.a[1:nz-2, 0:nx-3]*ux[1:nz-2, 0:nx-3] - conc.a[1:nz-2, 2:nx-1]*ux[1:nz-2, 2:nx-1] ) 
fluxz_cen = np.zeros((nz, nx))
fluxz_cen[1:nz-2, 1:nx-2] = 0.5 * ( conc.a[0:nz-3, 1:nx-2]*uz[0:nz-3, 1:nx-2] - conc.a[2:nz-1, 1:nx-2]*uz[2:nz-1, 1:nx-2] )

# define anti-diffusive flux; shape =  [nz, nx]
adfx = fluxx_cen - fluxx_up
adfz = fluxz_cen - fluxz_up

# max and min concentrations in region
xup = np.zeros((nz, nx)); xdo = np.zeros((nz, nx))
zup = np.zeros((nz, nx)); zdo = np.zeros((nz, nx))

for j in range(1, nx - 1):
    xup[1:nz-1, j] = max( np.max(conc.a[1:nz-1, j-1:j+1]), np.max(tau_up[1:nz-1, j-1:j+1]) )    # C
    xdo[1:nz-1, j] = min( np.min(conc.a[1:nz-1, j-1:j+1]), np.min(tau_up[1:nz-1, j-1:j+1]) )    # C
for i in range(1, nz - 1):
    zup[i, 1:nx-1] = max( np.max(conc.a[i-1:i+1, 1:nx-1]), np.max(tau_up[i-1:i+1, 1:nx-1]) )    # C
    zdo[i, 1:nx-1] = min( np.min(conc.a[i-1:i+1, 1:nx-1]), np.min(tau_up[i-1:i+1, 1:nx-1]) )    # C

# define pos/neg fluxes in x   
nfluxx = np.sign(adfx)*0.5*(np.sign(adfx) - 1)
pfluxx = np.sign(adfx)*0.5*(np.sign(adfx) + 1)
xpos = np.zeros((nz, nx)); xneg = np.zeros((nz, nx)) 
xpos[:, 0:nx - 1] = pfluxx[:, 0:nx - 1] * adfx[:, 0:nx - 1] - nfluxx[:, 1:nx - 0] * adfx[:, 1:nx - 0]
xneg[:, 0:nx - 1] = pfluxx[:, 1:nx - 0] * adfx[:, 1:nx - 0] - nfluxx[:, 0:nx - 1] * adfx[:, 0:nx - 1]

# define pos/neg fluxes in z 
zpos = np.zeros((nz, nx)); zneg = np.zeros((nz, nx))
nfluxz = np.sign(adfz)*0.5*(np.sign(adfz) - 1)
pfluxz = np.sign(adfz)*0.5*(np.sign(adfz) + 1)
zpos[0:nz - 1, :] = pfluxz[0:nz - 1, :] * adfz[0:nz - 1, :] - nfluxz[1:nz - 0, :] * adfz[1:nz - 0, :]
zneg[0:nz - 1, :] = pfluxz[1:nz - 0, :] * adfz[1:nz - 0, :] - nfluxz[0:nz - 1, :] * adfz[0:nz - 1, :]

# calculate the Zalesak parameter
zbetaup = (zup - tau_up) / zpos #* conc.dx/dt                                                 # C / (C m/s) * m/s = non dimensional
zbetado = (tau_up - zdo) / zneg #* conc.dx/dt
xbetaup = (xup - tau_up) / xpos #* conc.dz/dt
xbetado = (tau_up - xdo) / xneg #* conc.dz/dt
# remove nans and infs
zeros = np.zeros(np.shape(zbetaup))
idx = np.isnan(zbetaup)
idx = idx + np.isinf(zbetaup)
zbetaup[idx] = 0
idx = np.isnan(zbetado)
idx = idx + np.isinf(zbetado)
zbetado[idx] = 0
idx = np.isnan(xbetaup)
idx = idx + np.isinf(xbetaup)
xbetaup[idx] = 0
idx = np.isnan(xbetado)
idx = idx + np.isinf(xbetado)
xbetado[idx] = 0

# calculate zau, xau, zbu, xbu
# =one by default
zau = np.ones((nz, nx))
xau = np.ones((nz, nx))
zbu = np.ones((nz, nx))
xbu = np.ones((nz, nx))
# =zbetado if zbetado < 1
zau[zbetado < 1] = zbetado[zbetado < 1]                           # non dim and on the flux points
xau[xbetado < 1] = xbetado[xbetado < 1]                           # non dim and on the flux points
zbu[zbetaup < 1] = zbetaup[zbetaup < 1]                           # non dim and on the flux points
xbu[xbetaup < 1] = xbetaup[xbetaup < 1]                           # non dim and on the flux points
# shift zbetaup by one index
zbetaup[0:nz - 1, 0:nx - 2] = zbetaup[0:nz - 1, 1:nx - 1]
xbetaup[0:nz - 2, 0:nx - 1] = xbetaup[1:nz - 1, 0:nx - 1]
zbetado[0:nz - 1, 0:nx - 2] = zbetado[0:nz - 1, 1:nx - 1]
xbetado[0:nz - 2, 0:nx - 1] = xbetado[1:nz - 1, 0:nx - 1]
# =zbetaup if zbetaup[:, j + 1] < zbetado[:, j]
zau[zbetaup < zbetado] = zbetaup[zbetaup < zbetado]
xau[xbetaup < xbetado] = xbetaup[xbetaup < xbetado]
zbu[zbetado < zbetaup] = zbetaup[zbetado < zbetaup]
xbu[xbetado < xbetaup] = xbetaup[xbetado < xbetaup]
# set last column to zero since it's out of range
zau[:, nx - 1] = 0
xau[nz - 1, :] = 0
zbu[:, nx - 1] = 0
xbu[nz - 1, :] = 0

# calculate zcu & xcu
zcu = (0.5 + 0.5*np.sign(adfz))  
xcu = (0.5 + 0.5*np.sign(adfx))  

# calculate TVD flux in x and z
aaz = adfz * (zcu * zau + (1-zcu)*zbu)                                                   # C m/s on flux points
aax = adfx * (xcu * xau + (1-xcu)*xbu)

# final sol.
adv = np.zeros((nz, nx))
adv[1:nz-1, 1:nx-1] = tau_up[1:nz-1, 1:nx-1] +  (aax[1:nz-1, 0:nx-2] - aax[1:nz-1, 1:nx-1]) * conc.dx_i + (aaz[0:nz-2, 1:nx-1] - aaz[1:nz-1, 1:nx-1]) * conc.dz_i 





```
