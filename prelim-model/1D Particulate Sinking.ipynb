# Numerical Solution

```python
# finite-difference implementation of FTCS for linear advection.
#
# We are solving a_t = Q + S*a_z
#
# The FTCS discretization is: anew = aold + (C/2) (aold_{i+1} - aold_{i-1})
# 
# where S is the sinking velocity
#
# M. Zingale (2013-03-12)

import numpy
import pylab
%matplotlib inline

class FDgrid:

    def __init__(self, nz, ng, zmin=0.0, zmax=5000):

        self.zmin = zmin
        self.zmax = zmax
        self.ng = ng
        self.nz = nz

        # python is zero-based.  Make easy integers to know where the
        # real data lives
        self.ilo = ng
        self.ihi = ng+nz-1

        # physical coords
        self.dz = (zmax - zmin)/(nz-1)
        self.z = zmin + (numpy.arange(nz+2*ng)-ng)*self.dz

        # storage for the solution
        self.a = numpy.zeros((nz+2*ng), dtype=numpy.float64)

    def scratchArray(self):
        """ return a scratch array dimensioned for our grid """
        return numpy.zeros((self.nz + 2*self.ng), dtype=numpy.float64)

    def fillBCs(self):
        self.a[self.ilo] = 0
        self.a[self.ihi] = self.a[self.ihi - 1]


# create the grid
nz = 100
ng = 1
g = FDgrid(nz, ng)


# define the CFL, sink velocity, and reaction constant
S = 500        #m/yr
Q = 0.0267     #dpm/m^3

# time info
dt = 0.001
t = 0.0
tmax = 2.0*(g.zmax - g.zmin)/S            # time interval to reach bottom from surface
C = S*dt/g.dz


# initialize the data -- tophat
g.a[numpy.logical_and(g.z >= 0, g.z <= 5000)] = 1.0

ainit = g.a.copy()

# evolution loop
anew = g.scratchArray()

while (t < tmax):

    # fill the boundary conditions
    g.fillBCs()

    # loop over zones: note since we are periodic and both endpoints
    # are on the computational domain boundary, we don't have to
    # update both g.ilo and g.ihi -- we could set them equal instead.
    # But this is more general
    i = g.ilo
    while (i <= g.ihi):
        
        # FTCS
        #anew[i] = g.a[i] - 0.5*C*(g.a[i+1] - g.a[i-1]) + Q*dt           # C = S*dt/dz
        
         # upwind
        anew[i] = g.a[i] - C*(g.a[i] - g.a[i-1]) + Q*dt
        i += 1

    # store the updated solution
    g.a[:] = anew[:]

    t += dt

    
  
pylab.subplot(211)
pylab.title('Initial and Final [Th] Profiles')
pylab.plot(ainit[g.ilo:g.ihi], g.z[g.ilo:g.ihi])
pylab.gca().invert_yaxis()
pylab.ylabel('depth [m]')
pylab.xlim([0, 1.2])

pylab.subplot(2,1,2)
pylab.plot(abs(g.a[g.ilo:g.ihi]), g.z[g.ilo:g.ihi])
pylab.gca().invert_yaxis()
pylab.ylabel('depth [m]')
pylab.xlabel('[Th]')
pylab.xlim([0, 1.2])

pylab.savefig("fdadvect.png")
```

There is always production of the element tracer, preventing the distribution from approaching complete sedimentation. The distribution should be smooth.

```python
# steady state sol
g_a = Q*g.z/S

# plot results
#ax = pylab.subplots(1, 1, figsize = (10, 8))  
pylab.subplot(211)
pylab.title('Initial and Final [Th] Profiles')
pylab.plot(ainit[g.ilo:g.ihi], g.z[g.ilo:g.ihi])
pylab.gca().invert_yaxis()
pylab.ylabel('depth [m]')
pylab.xlim([0, 1.2])

pylab.subplot(212)
pylab.plot(g_a[g.ilo:g.ihi], g.z[g.ilo:g.ihi])
pylab.gca().invert_yaxis()
pylab.ylabel('depth [m]')
pylab.xlabel('[Th]')
pylab.xlim([0, 1.2])
```
