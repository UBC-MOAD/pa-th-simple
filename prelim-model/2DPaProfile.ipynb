Solves the [Pa] on a 2D 1e6m x 5e3m grid using an upwind advection scheme, and contains plots of [Pa] profile for a zero velocity field, a single cell velocity field, and a two cell velocity field.

# [Pa] for Corrected Two Cell Velocity Scheme

```python
# setup function adflow for Th 
reload(ThPa2D)
tmax = (0, 1, 2, 3, 5, 10, 20, 30, 60, 100)
t = tmax[0]
T = tmax[1]
V = 0.02/1000*365.*86400.  # first number in mm/s, corrections give needed value in m/yr ~ 630
xmin = 0
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nz_fine = 2*nz - 1
nx = 29
ng = 1
g = ThPa2D.FDgrid(nx, nz, ng)
h = ThPa2D.FDgrid(nx, nz, ng)

# chemistry
k_ad, k_de, Q = ThPa2D.k_sorp('Pa', xmin, xmax, zmin, zmax, nx, nz)

# initial profile
init = ThPa2D.plot_init(g, h, xmin, xmax, zmin, zmax, nx, nz, 'Pa')

# velocity
u, flowfig = ThPa2D.u_complex(xmin, xmax, zmin, zmax, nx, nz, V)
u_fine, flowfig = ThPa2D.u_complex(xmin, xmax, zmin, zmax, nx, nz_fine, V)
u, flowfig = ThPa2D.u_complex_c(u, u_fine, xmin, xmax, zmin, zmax, nx, nz)

# Th loop
g, h = ThPa2D.adflow(g, h, t, T, u, k_ad, k_de, Q, ThPa2D.upwind)

#plot results
meshTh = ThPa2D.plotprof(g, h, xmin, xmax, zmin, zmax, nx, nz, T, 'Pa')

# store results
DPa2vel = {}
PPa2vel = {}
DPa2vel['1tmax'] = g.a
PPa2vel['1tmax'] = h.a
```

Profiles show that the concentration in a water column is inversely proportional to downwelling. This is a result of upwelling currents effectively reducing the sinking velocity, leading to regions in which the Pa is suspended, or sinking very slowly. 

The dissolved and particulate phases appear to be identical except for a scaling factor.

Unfortunately, the solution does not reach a steady state before blowing up. 300 yrs is the last meaningful solution.
