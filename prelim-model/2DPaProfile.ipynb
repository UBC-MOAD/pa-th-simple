Solves the [Pa] on a 2D 1e6m x 5e3m grid using an upwind advection scheme, and contains plots of [Pa] profile for a zero velocity field, a single cell velocity field, and a two cell velocity field.

```python
from __future__ import division
import numpy
import pylab
import math
import matplotlib.pyplot as plt
from math import pi
import Th2D
%matplotlib inline
```

# [Pa] for Zero Velocity Field

```python
# set parameters

T = 1
V = 1/100
xmin = 1
xmax = 1000000
zmin = 0
zmax = 5000
nz = 20
nx = 20

# chemistry
[k_ad, k_de, Q] = Th2D.k_sorp('Pa', xmin, xmax, zmin, zmax, nx, nz)

# velocity
[u, flowfig] = Th2D.u_zero(xmin, xmax, zmin, zmax, nx, nz)

# [Th] loop
[flowfig, meshTh, DPa, PPa] = Th2D.adflow(T, V, u, nz, nx, k_ad, k_de, Q, flowfig, 'Pa')
```

```python
T = 2
[flowfig, meshTh, DPa, PPa] = Th2D.adflow(T, V, u, nz, nx, k_ad, k_de, Q, flowfig, 'Pa')
pylab.close(1)
```

```python
T = 3
[flowfig, meshTh, DPa, PPa] = Th2D.adflow(T, V, u, nz, nx, k_ad, k_de, Q, flowfig, 'Pa')
pylab.close(1)
```

```python
T = 5
[flowfig, meshTh, DPa, PPa] = Th2D.adflow(T, V, u, nz, nx, k_ad, k_de, Q, flowfig, 'Pa')
pylab.close(1)
```
