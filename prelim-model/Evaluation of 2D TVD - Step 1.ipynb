### Evaluation of 2D TVD : Step 1 (Determination that ux is spikey)

```python
from __future__ import division
import numpy as np
import math
import copy
import matplotlib.pyplot as plt
from math import pi
import ThPa2D
import flow
import plot
import profile_plots as pp
%matplotlib inline
```

Set up grid, uniform tracer and flow

```python
# Grid
xmin = 0
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nx = 30
ng = 1  # definition?
g = ThPa2D.Fgrid(nx, nz, ng)
h = ThPa2D.Fgrid(nx, nz, ng)

# Set tracer to 1 so advection terms aren't all zero.
g.a = 1*np.ones((nz,nx))

# Flow
V = 0.02/1000*365.*86400.
u = flow.onecell_cen(xmin, xmax, zmin, zmax, nx, nz, V)

# Plot Flow
init = plot.init(g.a, h.a, u, xmin, xmax, zmin, zmax, nx, nz, 'Th')
```

*So we have a uniform dissolved Th distribution of 1, a particular distribution of 0 and a single large eddy*

Now, following along from ThPa2D.py, adflow we set up for the TVD Scheme

```python
# extract the velocities
uz = u[0, :, :]
ux = u[1, :, :]

# upstream factors
sign_uz = np.sign(uz[:-1, :] + uz[1:, :])
sign_ux = np.sign(ux[:, :-1] + ux[:, 1:])

# and plot to check
fig, axs = plt.subplots(2,2,figsize=(7,7))
cmap = plt.get_cmap('bwr')

mesh = axs[0,0].pcolormesh(-uz, cmap=cmap)
axs[0,0].invert_yaxis()
axs[0,0].set_title('-uz')
cbar = fig.colorbar(mesh, ax=axs[0,0])

mesh = axs[0,1].pcolormesh(ux, cmap=cmap)
axs[0,1].invert_yaxis()
axs[0,1].set_title('ux')
cbar = fig.colorbar(mesh, ax=axs[0,1])

mesh = axs[1,0].pcolormesh(sign_uz, cmap=cmap)
axs[1,0].invert_yaxis()
axs[1,0].set_title('sign_uz')
cbar = fig.colorbar(mesh, ax=axs[1,0])

mesh = axs[1,1].pcolormesh(sign_ux, cmap=cmap)
axs[1,1].invert_yaxis()
axs[1,1].set_title('sign_ux')
cbar = fig.colorbar(mesh, ax=axs[1,1])
```

* Note that I have plotted these with upwards, upwards and with upwards velocity positive.
* Okay, the top two plots for uz and ux look good and consistent with a single vertical cell: We have rising water on the left side, "eastward" flow at the top, sinking water on the right side and "westward" flow at depth.
* Sign_uz looks fine.  This is the sign of uz (positive downward) and so is completely consistent with the plot above.
* Sign_ux has problems.  What is going on the far right?  Can we see something in the plot just above it on the far right?  Lets look closer.

```python
fig, axs = plt.subplots(1,2,figsize=(7,4))

axs[0].plot(ux[7,:])
axs[0].set_title('Ux, across z=7')
axs[1].plot(ux[14,:])
axs[1].set_title('Ux, across z=14')
```
