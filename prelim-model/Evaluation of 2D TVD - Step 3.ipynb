```python
from __future__ import division
import numpy as np
import math
import copy
import matplotlib.pyplot as plt
from math import pi
import flow
import plot
import profile_plots as pp
%matplotlib inline
```

```python
# Grid
xmin = 0
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nx = 28

# Flow Speed
V = 0.02/1000*365.*86400.
# define velocity on square grid, then scale onto rectangular grid. 
# calculate radius
a = zmax
b = zmax
x = np.linspace(-a/2, a/2, nx)
z = np.linspace(-b/2, b/2, nz)
[xx, zz] = np.meshgrid(x, z)
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)
rr = np.sqrt(xx**2 + zz**2)

# scale the flow in closer to centre of domain
scale = 0.9

# find idx
idx = rr < (a/2)*scale

# find uz (scaled in from bounadries and faded by Gaussian)
vsmall = 1e-12
uz = np.zeros([nz, nx])
uz[idx] = ( np.sin(2*pi*rr[idx] / (a*scale)) / (rr[idx]+vsmall) * xx[idx] 
    * np.exp(-4*rr[idx]**2/(a*scale)**2) )

# scale uz
uz = uz / np.max(uz) * V * zmax/xmax

# calculate the grid by grid divergence of uz
div = np.zeros((nz,nx))
div[1:nz-1,:] = - dx/dz* ( uz[2:nz,:] - uz[0:nz-2,:])

# set up ux
ux = np.zeros((nz,nx))

# integrate across to find ux
for j in range(2,nx):
    ux[:,j] = ux[:,j-2] + div[:,j-1]
```

```python
# look closely at ux at every depth level to check for spikes
fig, axs = plt.subplots(5,4,figsize=(15,15))
i = 0
for k in range(5):
    for l in range(4):
        axs[k,l].plot(ux[i,:])
        axs[k,l].set_title('ux across z = ' + str(i))
        i += 1
```

Since ux is still exhibiting wiggles in most of the profiles, try a larger decay rate:

Whenever the divergence is steep, ux contains spikes.

```python
# find uz (scaled in from bounadries and faded by Gaussian)
uz[idx] =  np.sin(2*pi*rr[idx] / (a*scale)) / (rr[idx]+vsmall) * xx[idx] * np.exp(-2*rr[idx]**2/(a/2*scale)**2) 

# scale uz
uz = uz / np.max(uz) * V * zmax/xmax

# calculate the grid by grid divergence of uz
div = np.zeros((nz,nx))
div[1:nz-1,:] = - dx/dz* ( uz[2:nz,:] - uz[0:nz-2,:])

# set up ux
ux = np.zeros((nz,nx))

# integrate across to find ux
for j in range(2,nx):
    ux[:,j] = ux[:,j-2] + div[:,j-1]
```

```python
# look closely at ux at every depth level to check for spikes
fig, axs = plt.subplots(5,4,figsize=(15,15))
i = 0
for k in range(5):
    for l in range(4):
        axs[k,l].plot(ux[i,:])
        axs[k,l].set_title('ux across z = ' + str(i))
        i += 1
```

```python
print(np.max(np.max(ux)))
print(ux[1, :])
print(ux[2, :])
```

Increasing the decay rate of the Gaussian by a factor of b decreases the amplitude of the oscillation in ux across z = 1, and z = 18 by a factor of 2^b.

Since the radius is never zero on an even grid, changing the decay rate changes the maximum value of ux.
