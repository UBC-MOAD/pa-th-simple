# Numerical Solution

```python
# finite-difference implementation of FTCS for linear advection.
#
# We are solving a_t = Q + S*a_z
#
# The FTCS discretization is: anew = aold + (C/2) (aold_{i+1} - aold_{i-1})
# 
# where S is the sinking velocity
#
# M. Zingale (2013-03-12)

import numpy
import pylab
%matplotlib inline

class FDgrid:

    def __init__(self, nz, ng, zmin=0.0, zmax=5000):

        self.zmin = zmin
        self.zmax = zmax
        self.ng = ng
        self.nz = nz

        # python is zero-based.  Make easy integers to know where the
        # real data lives
        self.ilo = ng
        self.ihi = ng+nz-1

        # physical coords
        self.dz = (zmax - zmin)/(nz-1)
        self.z = zmin + (numpy.arange(nz+2*ng)-ng)*self.dz

        # storage for the solution
        self.a = numpy.zeros((nz+2*ng), dtype=numpy.float64)

    def scratchArray(self):
        """ return a scratch array dimensioned for our grid """
        return numpy.zeros((self.nz + 2*self.ng), dtype=numpy.float64)

    def fillBCs(self):
        self.a[self.ilo] = 0
        self.a[self.ihi] = self.a[self.ihi - 1]


# create the grid
nz = 100
ng = 1
g = FDgrid(nz, ng)
h = FDgrid(nz, ng)

# define the CFL, sink velocity, and reaction constant
S = 300        #m/yr
Q = 0.0267     #dpm/m^3

# time info
dt = 0.001          #yr
t = 0.0
tmax = (g.zmax - g.zmin)/S    # max sinking time
C = S*dt/h.dz

# adsorption/desorption constants
k_ad = numpy.linspace(1, 1, len(g.z))
k_ad[251 <= g.z] = 0.75
k_ad[500 <= g.z] = 0.5
k_de = numpy.linspace(1.6, 1.6, len(g.z))

# initialize the data -- tophat
g.a[numpy.logical_and(g.z >= 0, g.z <= 5000)] = 0.5
h.a[numpy.logical_and(h.z >= 0, h.z <= 5000)] = 0.0

ainit = g.a.copy()
binit = h.a.copy()

# time evolution loop
anew = g.scratchArray()
bnew = h.scratchArray()

while (tmax <= 6*(g.zmax - g.zmin)/S ):
    
    while (t < tmax):               # loop through all dt on [0, tmax]

        # fill the boundary conditions
        g.fillBCs()
        h.fillBCs()

        # loop over zones: note since we are periodic and both endpoints
        # are on the computational domain boundary, we don't have to
        # update both g.ilo and g.ihi -- we could set them equal instead.
        # But this is more general
        i = g.ilo
        while (i <= g.ihi):         #loop through all depths

            # upstrea,
            anew[i] = g.a[i] + (Q - k_ad[i]*g.a[i] + k_de[i]*h.a[i]) * dt
            
            bnew[i] = h.a[i] - C*(h.a[i] - h.a[i-1]) + (k_ad[i]*g.a[i] -
                      k_de[i]*h.a[i]) * dt
            i += 1

        # store the updated solution
        g.a[:] = anew[:]
        h.a[:] = bnew[:]

        t += dt


    # initial dissolved and particulate
    pylab.figure(1, figsize = (16, 4))  
    pylab.plot(ainit[g.ilo:g.ihi], g.z[g.ilo:g.ihi])
    pylab.plot(binit[g.ilo:g.ihi], h.z[h.ilo:h.ihi])
    
    # final dissolved and particulate
    pylab.figure(2, figsize = (16, 8))
    pylab.subplot(211)
    pylab.plot(abs(g.a[g.ilo:g.ihi]), g.z[g.ilo:g.ihi])
    pylab.subplot(212)
    pylab.plot(abs(h.a[g.ilo:g.ihi]), h.z[h.ilo:h.ihi])
    
    # update tmax for next run
    tmax += (g.zmax - g.zmin)/S 

# label and scale plots
pylab.figure(2, figsize = (16, 8))
pylab.subplot(211)
pylab.ylabel('depth [m]')
pylab.xlim([0, 2.0])
pylab.title('Final Dissolved [Th] Profile')
pylab.gca().invert_yaxis()
pylab.legend(('tmax = 1', 'tmax = 2', 'tmax = 3', 'tmax = 4', 'tmax = 5'))
pylab.subplot(212)
pylab.xlim([0, 2.0])
pylab.title('Final Particulate [Th] Profile')
pylab.gca().invert_yaxis()
pylab.legend(('tmax = 1', 'tmax = 2', 'tmax = 3', 'tmax = 4', 'tmax = 5'))

pylab.figure(1, figsize = (16, 4))
pylab.gca().invert_yaxis()
pylab.title('Initial [Th] Profile')
pylab.legend(['dissolved', 'particulate'])
pylab.ylabel('depth [m]')
pylab.xlabel('[Th]')
pylab.xlim([-0.1, 1.0])
```

- Produce a 1D profile of [Th] in which dissolved and particulate phases are coupled by constant adsorption and desorption rates. 


- Code takes ~5 tmax to reach a linear steady state, where tmax = (g.zmax - g.zmin)/S, the time for a particle to sink from surface to floor (10 years with current parameters).


- The grid is 100 x 1.


- Before reaching a steady state, Th is uniformly distributed in depth, which is partly due to the production term Q, and partly due to the initial distribution. 


- It takes a time interval t = [0:tmax] for a particle to sink from an initial depth z = [0:zmax], which means after a time interval tmax, the initial distribution has fallen out. 


- Unintuitively, it takes longer than this to reach a steady state. The adsorption and desorption rates are significant enough that they extend the time elapsed before steady state. Since only the particulate phase of Th is subject to sinking, there could be a series of "stalls" in the sinking trajectory of a Th atom when is desorbs into the dissolved phase. This theory could be tested by changing the ad/de-sorption rates and comparing time to reach steady state.


- The steady state propogates from the surface to the floor. This is predominantly due to the initial distribution falling out at t = h/S, where h is the height of water above a given point, and S is sinking velocity, and partially due to a higher rate of adsorption near the surface.
