Rotate the circulation in x-z plane into y-z plane, and sum the two circulations.

```python
# define velocity on square grid, then scale onto rectangular grid. 
# overturning in x-z plane
a = zmax
b = zmax
c = zmax
x = np.linspace(-a/2, a/2, nx)
y = np.linspace(-b/2, b/2, ny)
z = np.linspace(-c/2, c/2, nz)
#zy = np.linspace(c/2, -c/2, nz)
# the order of mesh
[yy, zz, xx] = np.meshgrid(y, z, x)
#[yy, zzy, xx] = np.meshgrid(y, zy, x)
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)
dy = (ymax - ymin) / (ny - 1)
rr = np.sqrt(xx**2 + zz**2)
# scale the flow in closer to centre of domain
scale = 0.9
idx = rr < (a/2) * scale
uz = np.zeros([nz, ny, nx])
# apply a Gaussian to fade edges
uz[idx] = np.sin(2*pi*rr[idx] / (a*scale)) / rr[idx] * xx[idx] * np.exp(-2*rr[idx]**2/(a/2*scale)**2) 
# scale solution
uz = uz / np.max(uz) * V * zmax/xmax 
```

```python
# rotate by 90 degrees into y-z plane
rr = np.sqrt(yy**2 + zz**2)
# scale the flow in closer to centre of domain
idx = rr < (a/2) * scale
uz90 = np.zeros([nz, ny, nx])
# apply a Gaussian to fade edges
uz90[idx] = np.sin(2*pi*rr[idx] / (a*scale)) / rr[idx] * yy[idx] * np.exp(-2*rr[idx]**2/(a/2*scale)**2) 
# scale solution
uz90 = uz90 / np.max(uz90) * V * zmax/ymax 
```

```python
# define ux for uz in x-z
# vectorize in z, redefine u[j+1] with u[j-1] 
i = np.arange(1, nz - 1, 1, dtype = int)
j = 1
u = np.zeros((3, nz, ny, nx))
ux = u[1,:,:,:]
while j <= nx-2:

    ux[i,:,j+1] = ux[i,:,j-1] - dx/dz* ( uz[i+1,:,j] - uz[i-1,:,j])

    j += 1
    
# store result
u[0,:,:,:] = uz + uz90
```

```python
# define uy for uz in y-z
# vectorize in z, redefine u[j+1] with u[j-1] 
i = np.arange(1, nz - 1, 1, dtype = int)
j = 1
uy= u[2,:,:,:]
while j <= ny-2:

    uy[i,j+1,:] = uy[i,j-1,:] - dy/dz* ( uz90[i+1,j,:] - uz90[i-1,j,:])

    j += 1
```

# Rotate 2-cell velocity and add

```python
# define a grid that will produce downwelling
a = zmax
b = zmax
c = zmax

x = np.zeros(nx)
x[0:nx/2] = np.linspace(-a/2, a/2, nx/2)
x[nx/2:] = np.linspace(a/2, -a/2, nx/2)

y = np.zeros(nx)
y[0:ny/2] = np.linspace(-c/2, c/2, ny/2)
y[ny/2:] = np.linspace(c/2, -c/2, ny/2)

z = np.linspace(-b/2, b/2, nz)
[yy, zz, xx] = np.meshgrid(y, z, x)
zz[:, nx/2:,:] = - zz[:, nx/2:,:]  

rr = np.sqrt(xx**2 + zz**2)
# scale the flow in closer to centre of domain
scale = 0.9
idx = rr < (a/2) * scale
# apply a Gaussian to fade edges
uz = np.zeros((nz, ny, nx))
uz[idx] = np.sin(2*pi*rr[idx] / (a*scale)) / rr[idx] * xx[idx]* np.exp(-2*rr[idx]**2/(a/2*scale)**2) 
# scale solution
uz = uz / np.max(uz) * V * zmax/xmax 
# store solution in a tensor
u = np.zeros([3, nz, ny, nx])
ux = u[1,:,:,:]
# spatial step
dx = (xmax - xmin) / (nx - 1)
dy = (ymax - ymin) / (ny - 1)
dz = (zmax - zmin) / (nz - 1)

# vectorize in z, redefine u[j+1] with u[j-1] 
i = np.arange(1, nz - 1, 1, dtype = int)
j = 1
while j <= nx-2:

    ux[i,:,j+1] = ux[i,:,j-1] - dx/dz* ( uz[i+1,:,j] - uz[i-1,:,j])

    j += 1
```

```python
# define a grid that will produce downwelling
rr = np.sqrt(yy**2 + zz**2)
idx = rr < (a/2) * scale
# apply a Gaussian to fade edges
uz90 = np.zeros((nz, ny, nx))
uz90[idx] = np.sin(2*pi*rr[idx] / (a*scale)) / rr[idx] * yy[idx]* np.exp(-2*rr[idx]**2/(a/2*scale)**2) 
# scale solution
uz90 = uz90 / np.max(uz90) * V * zmax/ymax 
# store solution in a tensor
uy = u[2,:,:,:]
u[0,:,:,:] = uz90 + uz
# vectorize in z, redefine u[j+1] with u[j-1] 
i = np.arange(1, nz - 1, 1, dtype = int)
j = 1
while j <= ny-2:

    uy[i,j+1,:] = uy[i,j-1,:] - dy/dz* ( uz90[i + 1,j,:] - uz90[i-1,j,:])

    j += 1
```

Since uz and uz90 are perfectly symmetric with one another, uy and ux should always be symmetric, ie, they should always have the same sign. Problems arose in which the downwelling and upwelling occured away from convergent region of ux and uy. This was a result of using plt.gca().flip_yaxis. Anytime you use this command, you have to change the sign of the second argument in quiver. This is probably because python has already plotted the quiver plot, and as far as it knows, the command is simply to change the label on the axis. It would be nice if it also reversed the sign of the second argument in quiver, but it doesn't.

In order to apply the same y-flip to BOTH the imshow plot and the quiver plot, use extent = [xmin, xmax, ymax, ymin] BEFORE plotting the quiver. This way, python knows that when the second argument is < 0 , draw a vector pointing toward zero. 

It's good practise to flip axes before plotting in python to avoid this confusion in future.
