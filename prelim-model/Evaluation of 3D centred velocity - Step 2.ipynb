```python
# define velocity on square grid, then scale onto rectangular grid. 
# overturning in x-z plane
a = zmax
b = zmax
c = zmax
x = np.linspace(-a/2, a/2, nx)
y = np.linspace(-b/2, b/2, ny)
z = np.linspace(-c/2, c/2, nz)
# the order of mesh
[xx, zz, yy] = np.meshgrid(x, z, y)
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)
dy = (ymax - ymin) / (ny - 1)
rr = np.sqrt(xx**2 + zz**2)
# scale the flow in closer to centre of domain
scale = 0.9
idx = rr < (a/2) * scale
uz = np.zeros([nz, nx, ny])
# apply a Gaussian to fade edges
uz[idx] = np.sin(2*pi*rr[idx] / (a*scale)) / rr[idx] * xx[idx] * np.exp(-2*rr[idx]**2/(a/2*scale)**2) 
# scale solution
uz = uz / np.max(uz) * V * zmax/xmax 
```

```python
# rotate by 90 degrees into y-z plane
rr = np.sqrt(yy**2 + zz**2)
# scale the flow in closer to centre of domain
idx = rr < (a/2) * scale
uz90 = np.zeros([nz, nx, ny])
# apply a Gaussian to fade edges
uz90[idx] = np.sin(2*pi*rr[idx] / (a*scale)) / rr[idx] * yy[idx] * np.exp(-2*rr[idx]**2/(a/2*scale)**2) 
# scale solution
uz90 = uz90 / np.max(uz90) * V * zmax/ymax 
```

```python
# define ux for uz in x-z
# vectorize in z, redefine u[j+1] with u[j-1] 
i = np.arange(1, nz - 1, 1, dtype = int)
j = 1
u = np.zeros((3, nz, nx, ny))
ux = u[1,:,:,:]
while j <= nx-2:

    ux[i,j+1,:] = ux[i,j-1,:] - dx/dz* ( uz[i+1,j,:] - uz[i-1,j,:])

    j += 1
    
# store result
u[0,:,:,:] = uz
u[1,:,:,:] = ux
```

```python
# define uy for uz in y-z
# vectorize in z, redefine u[j+1] with u[j-1] 
i = np.arange(1, nz - 1, 1, dtype = int)
j = 1
uy= u[2,:,:,:]
while j <= ny-2:

    uy[i, :, j + 1] = uy[i, :, j - 1] - dy/dz* ( uz90[i+1,:,j] - uz90[i-1,:,j])

    j += 1
# store result
u[2,:,:,:] = uy
```

```python
# plot the different planar intersections
x = np.linspace(xmin, xmax, nx)
y = np.linspace(ymin, ymax, nx)
z = np.linspace(zmin, zmax, nz)
plt.subplots(1,3,figsize =(30,10))
plt.subplot(131)
plb.quiver(1e-3*x, z, u[1,:,:,15], -100*u[0,:,:,15], pivot = 'mid')
plb.gca().invert_yaxis()
plt.title('Velocity Field')
plt.xlabel('x [km]')
plt.ylabel('depth [m]')

plt.subplot(132)
plb.quiver(1e-3*y, z, u[2,:,15,:], -100*uz90[:,15,:], pivot = 'mid')
plb.gca().invert_yaxis()
plt.title('Velocity Field')
plt.xlabel('y [km]')
plt.ylabel('depth [m]')

plt.subplot(133)
plb.quiver(1e-3*x, 1e-3*y, u[2,10,:,:], u[1,10,:,:], pivot = 'mid')
plb.gca().invert_yaxis()
plt.title('Velocity Field')
plt.xlabel('x [km]')
plt.ylabel('y [km]')
```

# Rotate 2-cell velocity and add

```python
# define a grid that will produce downwelling
a = zmax
b = zmax
c = zmax

x = np.zeros(nx)
x[0:nx/2] = np.linspace(-a/2, a/2, nx/2)
x[nx/2:] = np.linspace(a/2, -a/2, nx/2)

y = np.zeros(nx)
y[0:nx/2] = np.linspace(-c/2, c/2, ny/2)
y[nx/2:] = np.linspace(c/2, -c/2, ny/2)

z = np.linspace(-b/2, b/2, nz)
[xx, zz, yy] = np.meshgrid(x, z, y)
zz[:, nx/2:,:] = - zz[:, nx/2:,:]  

rr = np.sqrt(xx**2 + zz**2)
# scale the flow in closer to centre of domain
scale = 0.9
idx = rr < (a/2) * scale
# apply a Gaussian to fade edges
uz = np.zeros((nz, nx, ny))
uz[idx] = -np.sin(2*pi*rr[idx] / (a*scale)) / rr[idx] * -xx[idx]* np.exp(-2*rr[idx]**2/(a/2*scale)**2) 
# scale solution
uz = uz / np.max(uz) * V * zmax/xmax 
# store solution in a tensor
u = np.zeros([3, nz, nx, ny])
ux = u[1,:,:,:]
u[0,:,:,:] = uz
# spatial step
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)

# vectorize in z, redefine u[j+1] with u[j-1] 
i = np.arange(1, nz - 1, 1, dtype = int)
j = 1
while j <= nx-2:

    ux[i, j + 1, :] = ux[i, j - 1, :] - dx/dz* ( uz[i + 1,j, :] - uz[i - 1, j, :])

    j += 1

# store solution
u[1,:,:,:] = ux
```

```python
# define a grid that will produce downwelling
rr = np.sqrt(yy**2 + zz**2)
# scale the flow in closer to centre of domain
scale = 0.9
idx = rr < (a/2) * scale
# apply a Gaussian to fade edges
uz90 = np.zeros((nz, nx, ny))
uz90[idx] = -np.sin(2*pi*rr[idx] / (a*scale)) / rr[idx] * -yy[idx]* np.exp(-2*rr[idx]**2/(a/2*scale)**2) 
# scale solution
uz90 = uz90 / np.max(uz90) * V * zmax/ymax 
# store solution in a tensor
uy = u[2,:,:,:]
u[0,:,:,:] = uz90
# spatial step
dy = (ymax - ymin) / (ny - 1)

# vectorize in z, redefine u[j+1] with u[j-1] 
i = np.arange(1, nz - 1, 1, dtype = int)
j = 1
while j <= ny-2:

    uy[i, :, j + 1] = uy[i, :, j - 1] - dy/dz* ( uz90[i + 1,:,j] - uz90[i - 1, :,j])

    j += 1

# store solution
u[2,:,:,:] = uy
```

```python
# plot the different planar intersections
x = np.linspace(xmin, xmax, nx)
y = np.linspace(ymin, ymax, nx)
z = np.linspace(zmin, zmax, nz)
plt.subplots(1,3,figsize =(30,10))
plt.subplot(131)
plb.quiver(1e-3*x, z, ux[:,:,0], -100*uz[:,:,0], pivot = 'mid')
plb.gca().invert_yaxis()
plt.title('Velocity Field')
plt.xlabel('x [km]')
plt.ylabel('depth [m]')

plt.subplot(132)
plb.quiver(1e-3*y, z, uy[:,0,:], -100*uz90[:,0,:], pivot = 'mid')
plb.gca().invert_yaxis()
plt.title('Velocity Field')
plt.xlabel('y [km]')
plt.ylabel('depth [m]')

plt.subplot(133)
plb.quiver(1e-3*x, 1e-3*y, uy[10,:,:], ux[10,:,:], pivot = 'mid')
plb.gca().invert_yaxis()
plt.title('Velocity Field')
plt.xlabel('x [km]')
plt.ylabel('y [km]')
```
