```python
from __future__ import division
import numpy as np
import math
import copy
import matplotlib.pyplot as plt
from math import pi
import flow
import plot
import ThPa2D
import profile_plots as pp
%matplotlib inline
```

```python
# Grid
xmin = 0
xmax = 1e6
zmin = 0
zmax = 5e3
nz = 20
nx = 28
V = 0.02/1000*365.*86400.
# define velocity on square grid, then scale onto rectangular grid. 
# calculate radius
a = zmax
b = zmax
x = np.linspace(-a/2, a/2, nx)
z = np.linspace(-b/2, b/2, nz)
[xx, zz] = np.meshgrid(x, z)
dx = (xmax - xmin) / (nx - 1)
dz = (zmax - zmin) / (nz - 1)
rr = np.sqrt(xx**2 + zz**2)
scale = 0.9
```

The velocity should be zero outside of the ellipse r^2 = (2x/a)^2 + (2z/b)^2:

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[0].pcolormesh(rr > a/2*scale)
axs[0].set_title('domain migrated inward from edges')
axs[0].grid(axis='both', linestyle='-')
axs[1].pcolormesh(rr > a/2)
axs[1].set_title('un-migrated domain')
axs[1].grid(axis='both', linestyle='-')
```

If the z-velocity is zero outside of the migrated domain, then the x velocity should be as well. However, multiplying by zero closer to the centre will increase the divergence more than multiplying by zero nearer the edges.

```python
# calculate ux according to previous corrections
u = flow.onecell_cen(xmin, xmax, zmin, zmax, nx, nz, V)
ux = u[1,:,:]
uz = u[0,:,:]

# zero out values outside unmigrated domain:
kill_osc = np.zeros((nz, nx))
ux[rr > a/2] = kill_osc[rr > a/2]
u[1,:,:] = ux

# check to see ux is still divergenceless
conc = ThPa2D.Fgrid(nx, nz, 1)
fig = plt.subplots(1,2,figsize=(15,5))
plt.subplot(121)
divplot = flow.divtest_cen(conc, u, nx, nz)
plt.title('divergence of u: ux(outside unmigrated domain) = 0')
# zero out values outside inward-migrated domain:
ux[rr > a/2*scale] = kill_osc[rr > a/2*scale]
u[1,:,:] = ux
plt.subplot(122)
divplot = flow.divtest_cen(conc, u, nx, nz)
plt.title('divergence of u: ux(inward migrated domain) = 0')
```

Since simply multiplying by zero outside a certain radius creates unwanted non-zero divergence terms, try cutting out oscillation by making the number of zeros in the edge entries of ux symmetric:

```python
# fiurst calculate u 
u = flow.onecell_cen(xmin, xmax, zmin, zmax, nx, nz, V)
ux = u[1,:,:]
idx = ux == 0
```

```python
plt.pcolormesh(idx)
```

```python
# now flip the orientation so you can multiply ux by kill_osc[idx] and make it symmetric 
idx1 = np.empty_like(idx)
i=0
while i <= nx-1:
    idx1[:, i] = idx[:, nx-1-i]
    i += 1
```

```python
plt.pcolormesh(idx1)
```

```python
# now multiply ux by zeroes at these entries
ux[idx1] = kill_osc[idx1]*ux[idx1]
# and plot ux at depth z
fig, axs = plt.subplots(5,4,figsize=(15,15))
i = 0
for k in range(5):
    for l in range(4):
        axs[k,l].plot(ux[i,:])
        axs[k,l].set_title('ux across z = ' + str(i))
        i += 1
```

```python
# test the divergence
u[1,:,:] = ux
divplot = flow.divtest_cen(conc, u, nx, nz)
plt.title('divergence of u: ux(inward migrated domain) = 0')
```

This is pretty good (at least, it is a better method than multiplying by zero outside given radius), but the non-zero divergence could still cause problems, even thought it is order 1e-4.
